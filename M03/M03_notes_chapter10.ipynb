# OOP with python

```python
# Empty quick class
class Cat():
    pass

c = Cat()
print(c)
```

```python
# Quick class with empty constructor
class Dog():
    def __init__(self):
        pass

d = Dog()
print(d)
```

```python
# Class with attribute
class Person():
    def __init__(self, name):
        self.name = name

p = Person('wyatt')
print(p.name)
```

```python
# Inheritance
class Car():
    def exclaim(self):
        print("Im a car!")

class Honda(Car):
    pass

car1 = Car()
car2 = Honda()

car1.exclaim()
car2.exclaim()
```

```python
# Method Override with inheritance
class Car():
    def exclaim(self):
        print("Im a car!")

class Honda(Car):
    def exclaim(self):
        print("Im a honda!")

car1 = Car()
car2 = Honda()

car1.exclaim()
car2.exclaim()
```

```python
# Super with inheritance
class Car():
    def __init__(self):
        self.var1: int = 1
        self.var2: int = 2

    def exclaim(self):
        print(self.var1)
        print(self.var2)

class Honda(Car):
    def __init__(self):
        super().__init__()
        self.var1: int = 3
        self.var2: int = 4

car1 = Car()
car2 = Honda()

car1.exclaim()
car2.exclaim()
```

```python
# Classes with hidden internal methods
class Car():
    def __init__(self):
        self.var1: int = 1
        self.var2: int = 2

    def exclaim(self):
        print(self.var1)
        print(self.var2)

class Honda(Car):
    def __init__(self):
        super().__init__()
        self.var1: int = 3
        self.var2: int = 4

car1 = Car()
car2 = Honda()

car1.exclaim()
car2.exclaim()
```

```python
# Computed values for get()
class Circle():
    def __init__(self, radius):
        self.radius = radius
    
    @property # This is called a decorator
    def diameter(self):
        return 2 * self.radius
    
circle1 = Circle(2)
print(circle1.radius)
print(circle1.diameter) # Didnt call circle1.diameter()
```

```python
# Static class variable
class A():
    count = 0
    def __init__(self):
        A.count += 1

    @classmethod
    def kids(cls):
        print(cls.count)

letter1 = A()
letter2 = A()
letter3 = A()
letter3.kids()
```

```python
# Static class method
class A():
    count = 0
    @classmethod
    def exclaim(cls):
        cls.count += 1
        print("All a's say the same")

letter1 = A()
letter2 = A()
letter3 = A()

letter1.exclaim()
letter2.exclaim()
letter3.exclaim()

print(letter1.count)
print(letter2.count)
print(letter3.count)
```
